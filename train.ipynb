```python
import warnings
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
import reverse_geocoder as rg
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from xgboost import XGBRegressor
import pickle
```

```python
train = pd.read_csv('new_train.csv')
test = pd.read_csv('new_test.csv')
```

```python
cv = ShuffleSplit(n_splits=10, random_state=0)
clf = XGBRegressor(n_estimators=2, max_depth=17, learning_rate=1)
scores = cross_val_score(clf, train.drop('score', axis=1), train['score'], cv=cv, scoring='neg_mean_absolute_error')
scores.mean()
```

```python
x_train, x_val, y_train, y_val = train_test_split(train.drop('score', axis=1), train['score'], test_size=0.2, random_state=42)
```

```python
clf = XGBRegressor(n_estimators=2, max_depth=17, learning_rate=1)
clf.fit(x_train, y_train)
pred6 = clf.predict(x_val)
mean_absolute_error(y_val, pred6)
```
